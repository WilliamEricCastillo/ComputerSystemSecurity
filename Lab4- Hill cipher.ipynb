```python
def cipher_encryption(plain, key):
    # Handle condition if the message length is odd.
    if len(plain) % 2:
        plain += "X"

    # Convert msg to matrices
    alphabet_string = string.ascii_uppercase

    plain_array = []
    for i in plain:
        for j in alphabet_string:
            if i == j:
                plain_array.append(alphabet_string.index(j))

    matrix_needed = int(len(plain_array) / 2)
    plain_matrix = np.zeros((matrix_needed, 2, 1))

    number = 1
    for k in range(matrix_needed):
        for m in range(2):
            plain_matrix[k][m] = plain_array[number - 1]
            number += 1

    plain_matrix = plain_matrix.reshape(matrix_needed, 2)

    # Convert key to 2x2
    key_array = []
    for i in key:
        for j in alphabet_string:
            if i == j:
                key_array.append(alphabet_string.index(j))

    key_matrix = np.zeros((2, 2))
    key_matrix[0] = [key_array[0], key_array[2]]
    key_matrix[1] = [key_array[1], key_array[3]]

    # print(key2d)
    print("Key Array Value: ")
    print(key_matrix)
    print("***************\n")

    # checking validity of the key
    # finding determinant
    determinant = key_matrix[0][0] * key_matrix[1][1] - key_matrix[0][1] * key_matrix[1][0]
    determinant = determinant % 26

    # finding multiplicative inverse
    mul_inv = -1
    for i in range(26):
        temp_inv = determinant * i
        if temp_inv % 26 == 1:
            mul_inv = i
            break
        else:
            continue

    if mul_inv == -1:
        print("Invalid key")
        sys.exit()

    # implementing steps to encrypt text
    encrypted_text = ""
    for g in range((len(plain_matrix))):
        p = g + 1
        m3 = np.dot(plain_matrix[g:p], key_matrix)
        m3 %= 26
        for e in range(2):
            for i in alphabet_string:
                if alphabet_string.index(i) == m3[0, e]:
                    encrypted_text += i

    print("Encrypted text: {}".format(encrypted_text))
    print("***************\n")

    return encrypted_text
```

```python
def cipher_decryption(ciphertext, key):
    # Handle condition if the message length is odd. 
    if len(ciphertext) % 2:
        ciphertext += "X"

    # Convert msg to matrices
    alphabet_string = string.ascii_uppercase

    ciphertext_array = []
    for i in ciphertext:
        for j in alphabet_string:
            if i == j:
                ciphertext_array.append(alphabet_string.index(j))

    matrix_needed = int(len(ciphertext_array) / 2)
    cipher_matrix = np.zeros((matrix_needed, 2, 1))

    number = 1
    for k in range(matrix_needed):
        for m in range(2):
            cipher_matrix[k][m] = ciphertext_array[number - 1]
            number += 1

    cipher_matrix = cipher_matrix.reshape(matrix_needed, 2)

    # Convert key to 2x2
    key_array = []
    for i in key:
        for j in alphabet_string:
            if i == j:
                key_array.append(alphabet_string.index(j))

    key_matrix = np.zeros((2, 2))
    key_matrix[0] = [key_array[0], key_array[2]]
    key_matrix[1] = [key_array[1], key_array[3]]

    # checking validity of the key
    determinant = key_matrix[0][0] * key_matrix[1][1] - key_matrix[0][1] * key_matrix[1][0]
    determinant = determinant % 26

    # finding multiplicative inverse
    mul_inv = -1
    for i in range(26):
        temp_inv = determinant * i
        if temp_inv % 26 == 1:
            mul_inv = i
            break
        else:
            continue

    if mul_inv == -1:
        print("Invalid key")
        sys.exit()

    # adjudicate matrix
    key_matrix[0][0], key_matrix[1][1] = key_matrix[1][1], key_matrix[0][0]

    # Change Matrix Signs
    key_matrix[0][1] *= -1
    key_matrix[1][0] *= -1

    key_matrix[0][1] = key_matrix[0][1] % 26
    key_matrix[1][0] = key_matrix[1][0] % 26

    for i in range(2):
        for j in range(2):
            key_matrix[i][j] *= mul_inv

    # modulo
    for i in range(2):
        for j in range(2):
            key_matrix[i][j] = key_matrix[i][j] % 26

    # cipher to plain

    decrypted_text = ""
    for g in range((len(cipher_matrix))):
        p = g + 1
        m3 = np.dot(cipher_matrix[g:p], key_matrix)
        m3 %= 26
        for e in range(2):
            for i in alphabet_string:
                if alphabet_string.index(i) == m3[0, e]:
                    decrypted_text += i

    if decrypted_text[len(decrypted_text)-1] == "X":
        decrypted_text = decrypted_text[:-1]

    print("Decrypted text: {}".format(decrypted_text))
```
